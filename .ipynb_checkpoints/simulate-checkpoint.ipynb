```python
import mbuild as mb

import foyer
from foyer import Forcefield as FF
import os
```

```python
zdol = mb.load('test_molecules/Zdol.mol2')
krytox = mb.load('test_molecules/Krytox.mol2')
```

```python
diglyme = mb.load('test_molecules/PFDiglyme.mol2')
diglyme.visualize()
```

```python
pfe = FF(forcefield_files='PFEs.xml')

diglyme_typed = pfe.apply(diglyme)

mb.formats.lammpsdata.write_lammpsdata(diglyme_typed, filename='diglyme', overwrite='True')
```

```python
pfe = FF(forcefield_files='PFEs.xml')

zdol_typed = pfe.apply(zdol)

mb.formats.lammpsdata.write_lammpsdata(zdol_typed, filename='diglyme', overwrite='True')
```

```python
system = mb.fill_box(compound=[diglyme], 
                     n_compounds=[256], 
                     box=[4.7, 4.7, 4.7])
```

```python
system.visualize()
```

```python
pfe = FF(forcefield_files='PFEs.xml')

system_typed = pfe.apply(system)

mb.formats.lammpsdata.write_lammpsdata(system_typed, filename='diglyme_box')
```
